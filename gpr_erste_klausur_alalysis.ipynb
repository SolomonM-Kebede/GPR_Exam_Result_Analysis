```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gpr_1 = pd.read_excel("erste Klausur gpr ergebnis.xltx")
gpr1_df = pd.DataFrame(gpr_1)
gpr1_df.head()
```

```python
new_index = range(1, len(gpr1_df)+1)
gpr1_df.index= new_index
pd.DataFrame(gpr1_df)
```

```python
gpr1_df['Note'] = gpr1_df['Note'].str.replace(',', '.')
gpr1_df['Note'] =pd.to_numeric(gpr1_df['Note'], errors='coerce')

pd.DataFrame(gpr1_df)
```

```python
class1 = []
class2 = []
class3 = []
class4 = []
for value in gpr1_df['Note']:
    if value >= 5.0:
        class1.append(value)
    elif 3.0<=value<5.0:
        class2.append(value)
    elif 2.0<= value < 3.0:
        class3.append(value)
    elif 1.0<= value < 2.0:
        class4.append(value)
    else:
        pass
total = [len(class1),len(class2),len(class3),len(class4)]
new_analysis = pd.DataFrame({'Nicht Bestanden': len(class1), 
                             'Bestanden': len(class2), 'good': len(class3), 
                             'very good': len(class4), 'Total':sum(total)}, index=[1])
new_analysis
```

```python
labels = ['Nicht Bestanden', 'Bestanden', 'good', 'very good']
plt.figure(figsize=(8, 8))  # Set the size of the figure
plt.pie(total, labels=labels, autopct='%1.1f%%', startangle=100)
plt.title('Pie Chart GPR1')  # Set the title of the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
