```python
new = ["Matrikelnr.", "Prozent", "Note"]
selected_data = df_gpr[new]
selected_data
```

```python

# Convert German-formatted numbers to standard format
selected_data['Note'] = selected_data['Note'].str.replace(',', '.')

# Convert the column to numeric with handling of non-numeric strings
selected_data['Note'] = pd.to_numeric(selected_data['Note'], errors='coerce')

# Display the DataFrame
print(selected_data)
```

```python
category1 =[]
category2 = []
category3 = []
category4 = []
for value in selected_data["Note"]:
    if value >= 5.0 :
        category1.append(value)
    elif 3<=value <5:
        category2.append(value)
    elif 2<=value <3:
        category3.append(value)
    elif 1<= value < 2: 
        category4.append(value)
```

```python
category = {'Nicht Bestanden':len(category1), 'Bestanden': len(category2), 'good': len(category3), 'Very Good': len(category4)}
analysis_gpr = pd.DataFrame(category, index=[1])
print(analysis_gpr)
```

```python
print(f'Die Anzahl von Studenten, die nicht bestanden {len(category1)}')
print(f'Die Anzahl von Studenten, die bestanden {len(category2)}')
print(f'Die Anzahl von Studenten, die "good" prädig bekommen {len(category3)}')
print(f'Die Anzahl von Studenten, die "Very Good" bekommen {len(category4)}')
```

```python
total = [len(category1),len(category2), len(category3), len(category4)]
labels = ['Nicht Bestanden', 'Bestanden', 'good', 'very good']
plt.figure(figsize=(6, 6))  # Set the size of the figure
plt.pie(total, labels=labels, autopct='%1.1f%%', startangle=100)
plt.title('Pie Chart GPR1')  # Set the title of the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
